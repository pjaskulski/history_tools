## SPACY

```python
import spacy
from spacy import displacy
```

```python
file = "chojnacka_1952.txt"
with open (file, "r", encoding='utf-8') as myfile:
    text = myfile.read()

nlp = spacy.load("pl_core_news_lg")
doc = nlp(text)
```

```python
displacy.render(doc, style="ent")
```

## CLARIN-PL

```python
import json
import os
import requests

user = "twoj.adres@email.com"
lpmn = 'any2txt|wcrft2|liner2({"model":"all"})'
url = "http://ws.clarin-pl.eu/nlprest2/base/process"

def process(file):
    with open (file, "r", encoding='utf-8') as myfile:
        doc = myfile.read()
        data = {'text': doc, 'lpmn': lpmn, 'user': user}        
        r = requests.post(url, data=json.dumps(data))
        if r.ok:
            return r.content.decode(encoding="utf-8")
        else:
            return str(r.status_code)    

file = "chojnacka_1952.txt"
print("Processing: " + file);        
data = process(file)
print(data)
```
